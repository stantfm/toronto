```python
import pandas as pd
```

```python
#Read Data from website and build dataframe

df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
headers = ['Postal Code','Borough','Neighbourhood']
df.columns = headers 
df.to_csv('newsource.txt')
df.head()
```

```python
#Remove columns when Borough = Not assigned
df = df.set_index('Borough')
df = df.drop('Not assigned', axis=0)
df.head()
```

```python
#Show shape of dataset

df.shape
```

```python
import pandas as pd

url="http://cocl.us/Geospatial_data"
df2=pd.read_csv(url)
df2.head(50)
```

```python
#Merge zip code data (=df2) into Toronto data (=df) using zip code as criteria:

inner_join = pd.merge(df,  
                      df2,  
                      on ='Postal Code',  
                      how ='inner')
inner_join
```
